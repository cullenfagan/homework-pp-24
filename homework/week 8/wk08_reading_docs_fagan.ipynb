````python
appelant_list = []
for target_file in target_files:
  with open(target_file, "r") as my_text:
    all_text = my_text.readlines()
    year = all_text[0].replace("Year: ", "").rstrip()
    print(year)
    cognition_related = all_text[2]
    if "not" in cognition_related:
      cognition_related = "False"
    else:
      cognition_related = "True"
    print(cognition_related)
    medical_condition = all_text[2]
    if "not" in medical_condition:
      medical_condition = "Not Specified"
    elif "Alzheimer" in medical_condition:
      medical_condition = "Alzheimer's"
    else:
      medical_condition = "Dementia"
    print(medical_condition)
    care_hours = all_text[4]
    if "12" in care_hours:
      care_hours = "half-day"
    elif "24" in care_hours:
      care_hours = "full-day"
    else:
      care_hours = "Not Specified"
    print(care_hours)
    print('***********')
    appelant_list.append({'Year': year, 'Cognition Related': cognition_related, 'Medical Condition': medical_condition, 'Care hours': care_hours})
````

````python
appelant_list
````

````python
df = pd.DataFrame(appelant_list)
df
````

````python
pd.DataFrame.to_csv(df, 'appelant_list.csv')
````
