```python
#import library
#what pulls down all the html from the server
import requests
```

```python
#requesting web content
url = "https://bestsellingalbums.org/decade/2010"
#scrape url website
response = requests.get(url)
```

```python
response.status_code
```

```python
response.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
soup.find(class_="album")
```

```python
albums = soup.find_all(class_="album")
albums[1]
```

```python
albums_list = []

for album in albums:
    album = album.get_text()
    albums_list.append(album)
    
albums_list
```
