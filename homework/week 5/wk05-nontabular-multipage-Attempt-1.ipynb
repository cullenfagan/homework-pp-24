Scrape the top 500 <a href="https://bestsellingalbums.org/decade/2010">best-selling albums of the 2010's</a>. Your data must include the following datapoints:

- Name of album
- Name of artist
- Number of albums sold 
- The link to the page that breaks down sales by country (found by clicking album title)

```python
#import library
#what pulls down all the html from the server
import requests
```

```python
#requesting web content
url = "https://bestsellingalbums.org/decade/2010"
#scrape url website
response = requests.get(url)
```

```python
response.status_code
```

```python
response.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
soup.find(class_="album")
```

```python
albums = soup.find_all(class_="album")
albums[1]
```

```python
albums_list = []

for album in albums:
    album = album.get_text()
    albums_list.append(album)
    
albums_list
```

```python
artists = soup.find_all(class_="artist")
artists[1]
```

```python
artists_list = []

for artist in artists:
    artist = artist.get_text()
    artists_list.append(artist)
    
artists_list
```

```python
sales = soup.find_all(class_="sales")
sales[1]
```

```python
sales_list = []

for sale in sales:
    number = sale.get_text().replace("Sales: ","").replace(",","")
    sales_list.append(number)
    
sales_list
```

```python
links_list = []

for album in albums:
    link = album.get("href")
    links_list.append(link)
    
links_list
```
